```python
from sklearn.neural_network import MLPRegressor
import pandas as pd
from joblib import dump, load
from os import path

import sys, os
import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")
    os.environ["PYTHONWARNINGS"] = "ignore" # Also affect subprocesses

REGRESSION_MODELS_FOLDER = 'automl/data/regression/models'
CLASSIFICATION_MODELS_FOLDER = 'automl/data/classification/models'
```

```python
regression_df = pd.read_csv('automl/data/regression/stats.csv')
regression_df
```

```python
X = regression_df.iloc[:,1:-19]
Y = regression_df.iloc[:,-19:]
```

```python
for model_name in Y:
    mlp = MLPRegressor(hidden_layer_sizes=(5, 10))
    mlp.fit(X, Y[model_name])
    dump(mlp, path.join(REGRESSION_MODELS_FOLDER, model_name))
```

```python
classification_df = pd.read_csv('automl/data/classification/stats.csv')
classification_df
```

```python
X = classification_df.iloc[:,1:-19]
Y = classification_df.iloc[:,-19:]
```

```python
for model_name in Y:
    mlp = MLPRegressor(hidden_layer_sizes=(5, 10))
    mlp.fit(X, Y[model_name])
    dump(mlp, path.join(CLASSIFICATION_MODELS_FOLDER, model_name))
```
